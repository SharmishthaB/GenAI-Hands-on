```python
from transformers import pipeline

# 1. Setup the Model
print("Loading 'Expert' Agent (DistilBERT)...")
qa_buddy = pipeline(
    "question-answering", 
    model="distilbert-base-cased-distilled-squad", 
    tokenizer="distilbert-base-cased-distilled-squad"
)

running = True  # This flag controls the entire app

while running:
    print("\n" + "="*50)
    print("ðŸ“– STEP 1: PASTE YOUR TEXTBOOK CHAPTER")
    print("="*50)
    context = input("Paste text here (or type 'exit' to quit): ").strip()

    if context.lower() == 'exit':
        print("Goodbye!")
        break

    # 3. Question Loop
    print("\n--- Context Loaded! Ready for Questions ---")
    print("(Options: Type your question | 'new' for new text | 'exit' to quit)")
    
    while True:
        question = input("\nYour Question: ").strip()

        if question.lower() == 'exit':
            print("Shutting down... Happy studying!")
            running = False  # Tell the outer loop to stop
            break           # Break the inner loop
        
        if question.lower() == 'new':
            print("\nResetting context...")
            break 
            
        # Run inference
        result = qa_buddy(question=question, context=context)
        
        print(f"\n [Answer]: {result['answer']}")
        print(f" [Confidence]: {round(result['score'] * 100, 2)}%")

print("\n[Status: Process Finished Cleanly]")
```
